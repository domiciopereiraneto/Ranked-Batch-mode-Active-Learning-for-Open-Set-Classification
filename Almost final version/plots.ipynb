```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
b25closed = np.load("Results/performancb25closed.npy")
b25open = np.load("Results/performancb25open.npy")
b50closed = np.load("Results/performancb50closed.npy")
b50open = np.load("Results/performancb50open.npy")
b100closed = np.load("Results/performancb100closed.npy")
b100open = np.load("Results/performancb100open.npy")
b250closed = np.load("Results/performancb250closed.npy")
b250open = np.load("Results/performancb250open.npy")
baseline_closed = np.load("Results/performancebaselineclosed.npy")
baseline_open = np.load("Results/performancebaselineopen.npy")
```

```python
b25x = np.arange(int(500/25)+1)*25
b50x = np.arange(int(500/50)+1)*50
b100x = np.arange(int(500/100)+1)*100
b250x = np.arange(int(500/250)+1)*250
```

```python
with plt.style.context('seaborn-white'):
    plt.figure(figsize=(10, 5))
    plt.title('Closed set micro-F1 score comparison by batch size')    
    plt.plot(b25x, b25closed, 'x-')
    plt.plot(b50x, b50closed, '^--')
    plt.plot(b100x, b100closed, 'o-.')
    plt.plot(b250x, b250closed, 's:')
    plt.plot([0, 500], np.repeat(baseline_closed, 2))
    plt.legend(['S=25','S=50','S=100','S=250','Baseline'], fontsize = 'large')
    plt.xlabel('Number of samples added')
    plt.ylabel('Micro-f1 score')
    plt.show()
```

```python
with plt.style.context('seaborn-white'):
    plt.figure(figsize=(10, 5))
    plt.title('Closed set micro-F1 score comparison by batch size')    
    plt.plot(b25x, b25open, 'x-')
    plt.plot(b50x, b50open, '^--')
    plt.plot(b100x, b100open, 'o-.')
    plt.plot(b250x, b250open, 's:')
    plt.plot([0, 500], np.repeat(baseline_open, 2))
    plt.legend(['S=25','S=50','S=100','S=250','Baseline'], fontsize = 'large')
    plt.xlabel('Number of samples added')
    plt.ylabel('Micro-f1 score')
    plt.show()
```
